# **Problem 1 TF-IDF**

## 📝**Problem Description**

Implement TF-IDF using using Python, Numpy, Pandas and whatever text cleaning library required.

The tf–idf is the product of two statistics, term frequency and inverse document frequency. There are various ways for determining the exact values of both statistics, you can use the following formulas.

### Term Frequency
$$tf_{t,d} = \log_{10}(count(t,d) +1)$$

* $tf_{t,d}$ is the frequency of the word t in the
document d

### Inverse Document Frequency
$$idf_t = \log_{10}(\frac{N}{df_t})$$

* $N$ is the total number of documents
* $df_t $ is the number of documents in which term t occurs

### TF-IDF
$$tf\text{-}idf_{t,d} = tf_{t,d} \times idf_t $$

Our goal is to get the TF-IDF vector for a given word.

## ✅**Requirements**

1. Function Definitions.
- Function 1 - `compute_tfidf_weights(train_docs)`
 - Input: A list of documents (each document is a list of words).
 - Output:
   - A TF matrix (how frequent each word appears in each document): each row corresponds to a document, each column to a word's TF value.
   - An IDF vector (how rare each word is across the documents): each entry corresponds to a word's IDF value.
   
- Function 2 - `word_tfidf_vector(word, tf_df, idf_df)`
 - Input: A word, the TF matrix (`tf_df`), and the IDF vector (`idf_df`).
 - Output: A TF-IDF vector for the input word across all documents.

2. Computation of TF matrix.
3. Computation of IDF vector.
4. Computation of the word TF-IDF vector for "cat".

````python
# ===========================
# === [Requirement 1: Function Implementations] ===
# ===========================

import numpy as np
import pandas as pd

# Function 1: Compute TF and IDF matrices, please complete the fucntion.
def compute_tfidf_weights(train_documents):
    # This function calculates term frequency–inverse document frequency (TF-IDF) weights.
    # It takes a list of documents. Each document is a list of words.
    # It returns:
    # 1. A DataFrame with term frequency values.
    # 2. A Series with inverse document frequency values.

    # Make a list of all unique words (vocabulary)
    vocabulary = set()
    for document in train_documents:
        for word in document:
            vocabulary.add(word)
    vocabulary = sorted(vocabulary)  # Sort the vocabulary for consistent order

    # Get the total number of documents
    total_number_of_documents = len(train_documents)

    # Prepare to store:
    # 1. Term frequency data for all documents
    # 2. Document frequency counts (how many documents contain each word)
    term_frequency_data = []
    document_frequency_counts = {}
    for word in vocabulary:
        document_frequency_counts[word] = 0

    # Go through each document
    for document in train_documents:
        # Count how many times each word appears in the document
        word_counts = {}
        for word in vocabulary:
            count = document.count(word)
            word_counts[word] = count

        # Make a list of term frequency values for this document
        term_frequency_row = []
        for word in vocabulary:
            # If the word appears, increase its document frequency count
            if word_counts[word] > 0:
                document_frequency_counts[word] += 1

            # Calculate term frequency using log scale: log10(count + 1)
            term_frequency_value = np.log10(word_counts[word] + 1)
            term_frequency_row.append(term_frequency_value)

        # Add this row to the full term frequency data
        term_frequency_data.append(term_frequency_row)

    # Create a DataFrame for term frequencies
    term_frequency_dataframe = pd.DataFrame(term_frequency_data, columns=vocabulary)

    # Calculate inverse document frequency values for each word
    inverse_document_frequency_values = []
    for word in vocabulary:
        number_of_documents_with_word = document_frequency_counts[word]

        # If word appears in at least one document, calculate IDF
        if number_of_documents_with_word > 0:
            inverse_document_frequency_value = np.log10(
                total_number_of_documents / number_of_documents_with_word
            )
        else:
            inverse_document_frequency_value = 0  # If never seen, IDF is 0

        inverse_document_frequency_values.append(inverse_document_frequency_value)

    # Create a Series for inverse document frequency values
    inverse_document_frequency_series = pd.Series(inverse_document_frequency_values, index=vocabulary)

    # Return the term frequency table and the inverse document frequency list
    return term_frequency_dataframe, inverse_document_frequency_series
````

````python
# Function 2: Compute word-level TF-IDF vector, please complete the fucntion.
def word_tfidf_vector(word, term_frequency_dataframe, inverse_document_frequency_series):
    # This function calculates the term frequency–inverse document frequency values for one word.
    # It tells us how important the word is in each document.

    # Parameters:
    # word: The word we are checking.
    # term_frequency_dataframe: A table that shows how often each word appears in each document.
    # Each row is a document. Each column is a word. Each value is the term frequency.
    # inverse_document_frequency_series: A list that shows how rare each word is across all documents.

    # Returns:
    # A list of term frequency–inverse document frequency values for the given word, one for each document.
    # If the word is not found, return a list of zeros (one for each document).

    # Check if the word is in both the term frequency table and the inverse document frequency list
    if word in term_frequency_dataframe.columns and word in inverse_document_frequency_series:
        # Multiply the term frequency values of the word by its inverse document frequency value
        return term_frequency_dataframe[word] * inverse_document_frequency_series[word]
    else:
        # If the word is not found, return a list of 0.0 values (same number as documents)
        return pd.Series([0.0] * len(term_frequency_dataframe))
````

````python

# ===========================
# === [Requirement 2: Computation of TF Matrix] ===
# ===========================

# Documents
train_docs = [
    ["the", "cat", "sat", "on", "the", "mat"],
    ["the", "dog", "sat", "on", "the", "log"],
    ["the", "cat", "chased", "the", "dog"]
]

# Call function to compute TF and IDF

# Display TF vector

# ===========================
# === [Requirement 3: Computation of IDF Vector] ===
# ===========================

term_frequency_dataframe, inverse_document_frequency_series = compute_tfidf_weights(train_docs)
print("=== Term Frequency (TF) Matrix ===")
print(term_frequency_dataframe.round(5))

# ===========================
# === [Requirement 4: Computation of Word TF-IDF Vector] ===
# ===========================

print("\n=== Inverse Document Frequency (IDF) Vector ===")
print(inverse_document_frequency_series.round(6))

# Compute TF-IDF vector for word "cat"
print("\n=== TF-IDF Vector for 'cat' ===")
print(word_tfidf_vector("cat", term_frequency_dataframe, inverse_document_frequency_series).round(6))
````

````python
# Load Dataset
import pandas as pd
train_data_url = "https://drive.google.com/uc?export=download&id=19JmVSOZ85vikn5aKna97aL5LM8KtG3T7"
test_data_url = "https://drive.google.com/uc?export=download&id=19EnwRfr6q5lzVB_UpJlGOG3IxgDhYVgP"

df_train = pd.read_csv(train_data_url, encoding='latin-1')[["OriginalTweet","Sentiment"]].rename(columns={'OriginalTweet': 'tweet', 'Sentiment': 'label'})
df_test = pd.read_csv(test_data_url, encoding='latin-1')[["OriginalTweet","Sentiment"]].rename(columns={'OriginalTweet': 'tweet', 'Sentiment': 'label'})

df_test
````

````python
# ==============================
# Import Necessary Libraries
# ==============================

import pandas as pd
import numpy as np
import string
import nltk
import torch
import torch.nn as neural_network  # Alias for torch.nn to clarify usage
import torch.optim as optimizer_library  # Alias for torch.optim
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder
import gensim.downloader as api
import torch.utils.data as data_utils

# Download necessary NLTK data files for tokenization and stopwords
nltk.download('punkt')
nltk.download('stopwords')

# ===========================================
# [Requirement 1: Preprocessing Tweets]
# ===========================================

def preprocess_tweet_text(tweet_text):
    """
    Preprocess tweet text by tokenizing and removing stopwords and punctuation.

    Parameters:
    tweet_text (str): The raw tweet string.

    Returns:
    list: List of cleaned tokens (words) from the tweet.

    Example:
    preprocess_tweet_text("Hello, world!") -> ['hello', 'world']
    """
    # Convert tweet to lowercase and tokenize into words
    tokens = word_tokenize(tweet_text.lower())

    # Create a set of English stopwords for quick lookup
    stop_words_set = set(stopwords.words('english'))

    # Remove stopwords and punctuation from tokens
    tokens = [word for word in tokens if word not in stop_words_set and word not in string.punctuation]

    return tokens

# Apply preprocessing function to training and test tweet data
# This creates a new column 'token_list' with cleaned tokens for each tweet
df_train["token_list"] = df_train["tweet"].apply(preprocess_tweet_text)
df_test["token_list"] = df_test["tweet"].apply(preprocess_tweet_text)

# ===========================================
# [Requirement 2: Generate Tweet Embeddings]
# ===========================================

# Load pre-trained GloVe embeddings trained on Twitter data (100-dimensional vectors)
embedding_model = api.load("glove-twitter-100")

def compute_average_embedding(tokens, embedding_model):
    """
    Compute the average embedding vector for a list of tokens.

    Parameters:
    tokens (list): List of token strings.
    embedding_model (gensim model): Pretrained word embeddings.

    Returns:
    numpy.ndarray: Average embedding vector of size equal to embedding dimension.
                    Returns zero vector if no tokens found in embedding model.

    Example:
    compute_average_embedding(['happy', 'day'], embedding_model) -> array of shape (100,)
    """
    # Collect embeddings only for tokens present in the model
    word_vectors = [embedding_model[word] for word in tokens if word in embedding_model]

    # Return mean vector if available; else return zero vector of correct size
    if word_vectors:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(embedding_model.vector_size)

# Compute average embedding vectors for each tweet in training and test sets
df_train["embedding_vector"] = df_train["token_list"].apply(lambda token_list: compute_average_embedding(token_list, embedding_model))
df_test["embedding_vector"] = df_test["token_list"].apply(lambda token_list: compute_average_embedding(token_list, embedding_model))

# Stack all embedding vectors into a single numpy array for model input
training_input_features = np.stack(df_train["embedding_vector"].values)
test_input_features = np.stack(df_test["embedding_vector"].values)

# Encode string labels as integers for ML models
label_encoder = LabelEncoder()
training_labels = label_encoder.fit_transform(df_train["label"])
test_labels = label_encoder.transform(df_test["label"])

# ===========================================
# [Requirement 3: Traditional Classifier]
# ===========================================

# Initialize and train logistic regression classifier on the tweet embeddings
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(training_input_features, training_labels)

# Predict labels on the test set
logistic_regression_predictions = logistic_regression_model.predict(test_input_features)

# Print evaluation metrics for logistic regression
print("===== Classifier 1 (Logistic Regression) =====")
print("Accuracy (Micro):", round(accuracy_score(test_labels, logistic_regression_predictions) * 100, 2), "%")
print("F1 Score (Micro):", round(f1_score(test_labels, logistic_regression_predictions, average='micro') * 100, 2), "%")
print("F1 Score (Macro):", round(f1_score(test_labels, logistic_regression_predictions, average='macro') * 100, 2), "%")

# ===========================================
# [Requirement 4: Neural Network Classifier]
# ===========================================

# Convert input features and labels to PyTorch tensors
training_input_tensor = torch.tensor(training_input_features, dtype=torch.float32)
test_input_tensor = torch.tensor(test_input_features, dtype=torch.float32)
training_label_tensor = torch.tensor(training_labels, dtype=torch.long)
test_label_tensor = torch.tensor(test_labels, dtype=torch.long)

# Create PyTorch dataset and dataloader for batching during training
training_dataset = data_utils.TensorDataset(training_input_tensor, training_label_tensor)
batch_size_value = 32
training_data_loader = data_utils.DataLoader(training_dataset, batch_size=batch_size_value, shuffle=True)

class FullyConnectedNeuralNetwork(neural_network.Module):
    """
    A simple feedforward neural network with two hidden layers and dropout.

    Args:
    input_dimension (int): Size of input feature vector.
    first_hidden_layer_dimension (int): Number of neurons in first hidden layer.
    second_hidden_layer_dimension (int): Number of neurons in second hidden layer.
    output_dimension (int): Number of output classes.
    dropout_probability (float): Dropout rate to prevent overfitting.
    """
    def __init__(self, input_dimension, first_hidden_layer_dimension, second_hidden_layer_dimension, output_dimension, dropout_probability=0.5):
        super().__init__()
        # First hidden layer with ReLU activation and dropout
        self.first_layer = neural_network.Linear(input_dimension, first_hidden_layer_dimension)
        self.first_activation = neural_network.ReLU()
        self.first_dropout = neural_network.Dropout(dropout_probability)

        # Second hidden layer with ReLU activation and dropout
        self.second_layer = neural_network.Linear(first_hidden_layer_dimension, second_hidden_layer_dimension)
        self.second_activation = neural_network.ReLU()
        self.second_dropout = neural_network.Dropout(dropout_probability)

        # Output layer with raw scores for each class
        self.output_layer = neural_network.Linear(second_hidden_layer_dimension, output_dimension)

    def forward(self, input_data):
        """
        Forward pass through the network.

        Parameters:
        input_data (torch.Tensor): Input feature tensor.

        Returns:
        torch.Tensor: Output logits (raw class scores).
        """
        # Pass input through first hidden layer, activation, and dropout
        input_data = self.first_dropout(self.first_activation(self.first_layer(input_data)))
        # Pass result through second hidden layer, activation, and dropout
        input_data = self.second_dropout(self.second_activation(self.second_layer(input_data)))
        # Compute output layer values
        return self.output_layer(input_data)

# Set model hyperparameters
input_dimension = training_input_features.shape[1]  # e.g., 100 for GloVe embeddings
first_hidden_layer_dimension = 256
second_hidden_layer_dimension = 128
output_dimension = len(np.unique(training_labels))  # Number of classes
dropout_probability = 0.5
learning_rate_value = 0.001
number_of_training_epochs = 30

# Instantiate the neural network model
neural_network_model = FullyConnectedNeuralNetwork(
    input_dimension,
    first_hidden_layer_dimension,
    second_hidden_layer_dimension,
    output_dimension,
    dropout_probability
)

# Define the loss function for classification (Cross-Entropy)
loss_function = neural_network.CrossEntropyLoss()

# Define the optimizer (Adam) to update model weights
model_optimizer = optimizer_library.Adam(neural_network_model.parameters(), lr=learning_rate_value)

def train_neural_network(model, loss_function, model_optimizer, data_loader, number_of_epochs):
    """
    Train the neural network model.

    Parameters:
    model (torch.nn.Module): The neural network to train.
    loss_function (callable): Loss function to minimize.
    model_optimizer (torch.optim.Optimizer): Optimizer for updating weights.
    data_loader (DataLoader): DataLoader providing batches of training data.
    number_of_epochs (int): Number of times to iterate over the training data.

    Returns:
    None
    """
    model.train()  # Set model to training mode

    for epoch_index in range(number_of_epochs):
        total_epoch_loss = 0

        # Iterate over batches of data
        for batch_features, batch_labels in data_loader:
            model_optimizer.zero_grad()  # Clear previous gradients
            model_output = model(batch_features)  # Forward pass
            batch_loss = loss_function(model_output, batch_labels)  # Compute loss
            batch_loss.backward()  # Backpropagation
            model_optimizer.step()  # Update weights

            total_epoch_loss += batch_loss.item()

        # Calculate average loss for the epoch
        average_loss = total_epoch_loss / len(data_loader)
        print(f"Epoch [{epoch_index + 1}/{number_of_epochs}], Loss: {average_loss:.4f}")

def evaluate_neural_network(model, test_features_tensor, actual_test_labels):
    """
    Evaluate the trained neural network on test data.

    Parameters:
    model (torch.nn.Module): Trained neural network.
    test_features_tensor (torch.Tensor): Test input features.
    actual_test_labels (np.ndarray): True labels for test data.

    Returns:
    None
    """
    model.eval()  # Set model to evaluation mode
    with torch.no_grad():
        # Get model predictions
        model_predictions = model(test_features_tensor)
        # Choose the class with the highest score
        predicted_labels = torch.argmax(model_predictions, dim=1).cpu().numpy()

    # Print evaluation metrics for neural network classifier
    print("\n===== Classifier 2 (Neural Network) =====")
    print("Accuracy (Micro):", round(accuracy_score(actual_test_labels, predicted_labels) * 100, 2), "%")
    print("F1 Score (Micro):", round(f1_score(actual_test_labels, predicted_labels, average='micro') * 100, 2), "%")
    print("F1 Score (Macro):", round(f1_score(actual_test_labels, predicted_labels, average='macro') * 100, 2), "%")

# Train the neural network with training data
train_neural_network(neural_network_model, loss_function, model_optimizer, training_data_loader, number_of_training_epochs)

# Evaluate neural network on test data
evaluate_neural_network(neural_network_model, test_input_tensor, test_labels)

# ===========================================
# [Requirement 5: Print Discussion]
# ===========================================

# Summary of results and comparison between classifiers
print("The neural network performed better than logistic regression.")
print("Its accuracy was 48%, while logistic regression had 43.6%.")
print("The F1 score (micro) was also higher: 48% vs 43.6%.")
print("The F1 score (macro) was better too: 49.2% vs 43.92%.")
print("The neural network learns complex patterns better, which helps improve performance.")
````

````python
# Helper functions
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples[f"upos"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            # Special tokens have a word id that is None. We set the label to -100 so they are automatically
            # ignored in the loss function.
            if word_idx is None:
                label_ids.append(-100)
            # We set the label for the first token of each word.
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            # For the other tokens in a word, we set the label to either the current label or -100, depending on
            # the label_all_tokens flag.
            else:
                label_ids.append(label[word_idx] if label_all_tokens else -100)
            previous_word_idx = word_idx

        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    # Masking ignored indices (e.g., -100) used during training
    true_labels = [l for p, l in zip(preds.flatten(), labels.flatten()) if l != -100]
    pred_labels = [p for p, l in zip(preds.flatten(), labels.flatten()) if l != -100]

    return {
        "accuracy": accuracy_score(true_labels, pred_labels),
        "f1": f1_score(true_labels, pred_labels, average='macro'),
        "precision": precision_score(true_labels, pred_labels, average='macro'),
        "recall": recall_score(true_labels, pred_labels, average='macro'),
    }
````

````python
import numpy as np
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForTokenClassification,
    DataCollatorForTokenClassification,
    TrainingArguments,
    Trainer
)

import os

# --- !!! ADD THIS LINE AT THE TOP !!! ---
os.environ["WANDB_DISABLED"] = "true"
# ---

# ======================
# === [Requirement 1: Data Loading & Preprocessing Function Implementation] ===
# ======================

# Configuration
MODEL_CHECKPOINT = "distilbert-base-cased"  # Using a lightweight version of BERT
DATASET_NAME = "universal_dependencies"     # Dataset with POS tags and dependency info
DATASET_SUBSET = "en_ewt"                   # English Web Treebank subset

# Load the dataset
# Contains train, validation, and test splits with sentence-level annotations
raw_datasets = load_dataset(DATASET_NAME, DATASET_SUBSET)
print(raw_datasets)

# Get label list (Universal POS tags)
# Universal POS tags include tags like NOUN, VERB, ADJ, etc.
label_list = raw_datasets["train"].features[f"upos"].feature.names
label_to_id = {label: i for i, label in enumerate(label_list)}  # Map each POS tag to a unique ID

# Load tokenizer
# Tokenizer splits text into subwords to match BERT input format
tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)

# Tokenize and align labels
# Applies tokenization to each sentence and aligns original word-level labels with tokenized output
label_all_tokens = False
tokenized_datasets = raw_datasets.map(tokenize_and_align_labels, batched=True)

# ======================
# === [Requirement 2: Label Alignment & Tokenization Logic] ===
# ======================

# Load pre-trained BERT model for token classification (POS tagging)
# Sets the number of output classes equal to the number of POS tags
model = AutoModelForTokenClassification.from_pretrained(
    MODEL_CHECKPOINT, num_labels=len(label_list)
)

# Data collator dynamically pads sequences during training/evaluation
# Ensures all samples in a batch have the same length
data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)

# ======================
# === [Requirement 3: Training Setup & Execution] ===
# === You need to test on the test set, hint: compute_metrics(trainer.predict(tokenized_datasets["test"])) ===
# ======================

# Define training arguments
# Includes training config like batch size, learning rate, number of epochs, etc.
training_args = TrainingArguments(
    output_dir=f"test-pos",
    eval_strategy="epoch",               # Evaluate at the end of each epoch
    learning_rate=2e-5,                  # Standard learning rate for fine-tuning
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,                   # Helps regularize the model to avoid overfitting
    report_to="none"                     # Disable logging to external services
)

# Trainer handles training, evaluation, and prediction
# Uses model, tokenizer, data, collator, and compute_metrics function
trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# Train the model
# Fine-tunes BERT on POS tagging task using the training data
trainer.train()

# Evaluate on test set
# Makes predictions on test data and computes performance metrics (e.g., accuracy, F1)
print("\nTest dataset metrics:")
results = trainer.predict(tokenized_datasets["test"])
metrics = compute_metrics(results)
print(metrics)
````
